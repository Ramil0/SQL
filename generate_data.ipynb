# Создание базы данных

```python
import sqlite3
import random 

conn = sqlite3.connect('metro.db')
conn.execute('PRAGMA foreign_keys = ON')
```

```python
c = conn.cursor()
#create TicketType
conn.execute('''CREATE TABLE TicketType
             (Name Varchar(100) PRIMARY KEY, 
             Priveleges Varchar(100) NOT NULL, 
             PeriodofValidity INTERGER)''')
#create Manager
conn.execute('''CREATE TABLE Manager
             (ManagerID Varchar(100) PRIMARY KEY, 
             Surname Varchar(100) NOT NULL, 
             Name Varchar(100) NOT NULL, 
             Patronymic Varchar(100))''')
#create Cashier
conn.execute('''CREATE TABLE Cashier
             (CashierID Varchar(100) PRIMARY KEY, 
             Surname Varchar(100) NOT NULL, 
             Name Varchar(100) NOT NULL, 
             Patronymic Varchar(100), 
             Experience INTEGER NOT NULL,
             ManagerID Varchar(100) NOT NULL, 
             FOREIGN KEY(ManagerID) REFERENCES Manager(ManagerID) ON UPDATE CASCADE ON DELETE CASCADE)''')
#create Route
conn.execute('''CREATE TABLE Route
            (RouteID Varchar(100) PRIMARY KEY)''')
#create Station
conn.execute('''CREATE TABLE Station (
            StationID Varchar(100) PRIMARY KEY,
            StationName Varchar(100) NOT NULL)''')
#create Train
conn.execute('''CREATE TABLE Train (TrainID Varchar(100) PRIMARY KEY,
            SessionBegining Varchar(100) NOT NULL,
            SessionEnd Varchar(100) NOT NULL,
            RouteID Varchar(100) NOT NULL,
            FOREIGN KEY(RouteID) REFERENCES Route(RouteID) ON UPDATE CASCADE ON DELETE CASCADE)''')
#create Passenger
conn.execute('''CREATE TABLE Passenger
             (PassengerID Varchar(100) PRIMARY KEY, 
             Surname Varchar(100) NOT NULL, 
             Name Varchar(100) NOT NULL, 
             Patronymic Varchar(100), 
             TrainID Varchar(100) NOT NULL, 
             FOREIGN KEY(TrainID) REFERENCES Train(TrainID) ON UPDATE CASCADE ON DELETE CASCADE)''')
#create Drivers
conn.execute('''CREATE TABLE Driver (DriverID text PRIMARY KEY,
            Surname Varchar(100) NOT NULL,
            Name Varchar(100) NOT NULL,
            Patronymic Varchar(100),
            Experience INTEGER NOT NULL,
            SessionBegining Varchar(100) NOT NULL,
            SessionEnd Varchar(100) NOT NULL,
            TrainID Varchar(100) NOT NULL,
            ManagerID Varchar(100) NOT NULL,
            FOREIGN KEY(TrainID) REFERENCES Train(TrainID) ON UPDATE CASCADE ON DELETE CASCADE, 
            FOREIGN KEY(ManagerID) REFERENCES Manager(ManagerID) ON UPDATE CASCADE ON DELETE CASCADE)''')
#create RouteToStation
conn.execute('''CREATE TABLE RouteToStation
             (ID Varchar(100) PRIMARY KEY, 
             OrderNumber INTEGER NOT NULL,
             RouteID Varchar(100) NOT NULL,
             StationID Varchar(100) NOT NULL,
             FOREIGN KEY(RouteID) REFERENCES Route(RouteID) ON UPDATE CASCADE ON DELETE CASCADE,
             FOREIGN KEY(StationID) REFERENCES Station(StationID) ON UPDATE CASCADE ON DELETE CASCADE)''')
#create StationToStation
conn.execute('''CREATE TABLE StationToStation
             (ID Varchar(100) PRIMARY KEY,
             FirstStation Varchar(100) NOT NULL,
             SecondStation Varchar(100) NOT NULL,
             FOREIGN KEY(FirstStation) REFERENCES Station(StationID) ON UPDATE CASCADE ON DELETE CASCADE,
             FOREIGN KEY(SecondStation) REFERENCES Station(StationID) ON UPDATE CASCADE ON DELETE CASCADE)''')
#create TicketMachine
conn.execute('''CREATE TABLE TicketMachine
             (MachineID Varchar(100) PRIMARY KEY, 
             StationID Varchar(100) NOT NULL,
             FOREIGN KEY(StationID) REFERENCES Station(StationID) ON UPDATE CASCADE ON DELETE CASCADE)''')
#create Ticket
conn.execute('''CREATE TABLE Ticket
             (TicketID Varchar(100) PRIMARY KEY, 
             NumberOfTrips INTEGER NOT NULL, 
             TypeOfTicket Varchar(100) NOT NULL,
             PassengerID Varchar(100) NOT NULL,
             StationID Varchar(100) NOT NULL,
             CashierID Varchar(100),
             MachineID Varchar(100),
             FOREIGN KEY(TypeOfTicket) REFERENCES TicketType(Name) ON UPDATE CASCADE ON DELETE CASCADE,
             FOREIGN KEY(PassengerID) REFERENCES Passenger(PassengerID) ON UPDATE CASCADE ON DELETE CASCADE,
             FOREIGN KEY(StationID) REFERENCES Station(StationID) ON UPDATE CASCADE ON DELETE CASCADE,
             FOREIGN KEY(CashierID) REFERENCES Cashier(CashierID) ON UPDATE CASCADE ON DELETE CASCADE,
             FOREIGN KEY(MachineID) REFERENCES TicketMachine(MachineID) ON UPDATE CASCADE ON DELETE CASCADE)''')
```
